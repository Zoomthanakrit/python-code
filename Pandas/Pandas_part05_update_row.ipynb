```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
#Replace column name
df.columns = ['first_name', 'last_name', 'email']
df
```

```python
#Use for upper in each column
df.columns = [x.upper() for x in df.columns]
df
```

```python
#Replace name in column name
df.columns = df.columns.str.replace('_', ' ')
df
```

```python
#Replace name in function
cols = {"FIRST NAME":"first", "LAST NAME":"last"}
df.rename(columns=cols, inplace=True)
df
```

```python
#Replace value
df.loc[2, 'last'] = "Smith"
df
```

```python
#Replace value one point
df.at[2, 'last'] = "Doe"
df
```

```python
#Lower all column
df['EMAIL'] = df['EMAIL'].str.lower()
df
```

```python
#Count character in each row
df['EMAIL'].apply(len)
```

```python
#Use function for apply
def email_update(email):
    return email.upper()

#Method1
df['EMAIL'].apply(email_update)

#Method2
df['EMAIL'].apply(lambda x: x.lower())
```

```python
#Count each series row | column
df.apply(len, axis=1)
```

```python
#Give the min in each column

#Method1
df.apply(pd.Series.min)

#Method2
df.apply(lambda x:x.min())
```

```python
#Applymap
df.applymap(len)
```

```python
#Applymap for all
df.applymap(str.lower)
df
```

```python
#Replace as per instruction but other keep the same
df['first'] = df['first'].replace({'Corey':'Chris', 'Jane': 'Mary'})

#Replace as per instruction but other from code will be none
df['first'] = df['first'].map({'Corey':'Chris', 'Jane': 'Mary'})
df
```
