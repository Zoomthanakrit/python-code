```python
#Group by is similar to filter
employ_group = df.groupby(['Employment'])
employ_group.get_group('Employed')
```

```python
#Groupby method 1
filt = df['Employment'] == 'Employed'
df.loc[filt]['EdLevel'].value_counts()
```

```python
#Groupby method 2
employ_group['EdLevel'].value_counts().loc['Employed']
```

```python
# Groupby find median
# employ_group['ConvertedCompYearly'].median().loc['Employed']
employ_group['ConvertedCompYearly'].median()
```

```python
#Groupby 2 condition
employ_group['ConvertedCompYearly'].agg(['median', 'mean'])
```

```python
#Sum specific
filt = df['Employment'] == 'Employed'
df.loc[filt]['AIAgentExternal'].str.contains('Claude').sum()
```

```python
#Use apply in groupby
employ_group['AIAgentExternal'].apply(lambda s: s.str.contains('Claude').sum())
```

```python
#
employ_percentage = df['Employment'].value_counts()
employ_percentage
```

```python
#
employ_claude = employ_group['AIAgentExternal'].apply(lambda s: s.str.contains('Claude').sum())
employ_claude
```

```python
#Concat new data
python_df = pd.concat([employ_percentage,employ_claude], axis='columns', sort=False)
python_df
```

```python
#rename
python_df.rename(columns={'count': 'No.of employee', 'AIAgentExternal': 'No. of use Claude'}, inplace=True)
python_df
```

```python
python_df['Percentage python'] = (python_df['No. of use Claude'] / python_df['No.of employee']) * 100
python_df
```

```python
python_df.sort_values(by='Percentage python', ascending=False)
```
