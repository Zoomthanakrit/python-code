```python
import re
#look up word that match
re.search("py", "python")
```

```python
#In case don't match
re.search("py", "pineapple")
```

```python
#Look up multiple locations
re.findall("py", "python pycharm pineapple")
```

```python
#wildcard
## end with at
re.findall('.at','hat cat at car')
```

```python
#have s or o
re.findall('so*', 'so sooooo s soooo')
```

```python
#start with so
re.findall('so+', 'so sooooo s soooo')
```

```python
#[] can be any smaller or capital
re.findall("[Pp]y", "python Pycharm pineapple")
```

```python
#select data that have py in word
#a-z
#A-Z
#0-9
#
#\. for only special character
re.findall("[Pp]y[a-zA-Z]*", "python Pycharm pineapple")
```

```python
re.findall('[a-zA-Z0-9-_]*@[a-zA-Z]*.[a-zA-Z]*', "Zoomthanakrit@gmail.com zoommiiie@xlo.com")
```
