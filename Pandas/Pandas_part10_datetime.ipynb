```python
#format sine download
d_parser = lambda x: pd.to_datetime(x, format='%Y-%m-%d %I-%p')
df = pd.read_csv('https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/refs/heads/master/Python/Pandas/10-Datetime-Timeseries/ETH_1h.csv', parse_dates=['Date'], date_parser=d_parser)

df.head()
```

```python
#Add dayname
df['Dayofweek'] = df['Date'].dt.day_name()
df
```

```python
df['Date'].min()
```

```python
df['Date'].max()
```

```python
#time range
df['Date'].max() - df['Date'].min()
```

```python
#Time range
# filt = (df['Date'] >= '2019') < (df['Date'] < '2020')
filt = (df['Date'] >= pd.to_datetime('2019-01-01')) < (df['Date'] < pd.to_datetime('2019-12-31'))
df.loc[filt]
```

```python
df.set_index('Date', inplace=True)
```

```python
df.sort_index(inplace=True)
```

```python
df['2020-01':'2020-02']['Close'].mean()
```

```python
df.loc['2020-01-01']['High'].max()
```

```python
#Max in a day
highs = df['High'].resample('D').max()
highs['2020-01-01']
```

```python
#Resample as many condition
df.resample('W').agg({'Close': 'mean', 'High': 'max', 'Low': 'min', 'Volume': 'sum'})
```
