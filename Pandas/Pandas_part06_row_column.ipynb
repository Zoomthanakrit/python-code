```python
people = {
    "first": ["Corey", 'Jane', 'John'], 
    "last": ["Schafer", 'Doe', 'Doe'], 
    "email": ["CoreyMSchafer@gmail.com", 'JaneDoe@email.com', 'JohnDoe@email.com']
}
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
#Create column
df['full_name'] = df['first'] + ' ' + df['last']
df['fullname'] = df['first'] + ' ' + df['last']
df
```

```python
#Remove column
df = df.drop(columns = ['full_name'])
df
```

```python
#Expand is set to column
df['fullname'].str.split(' ', expand = True)
```

```python
#Separate create new column
df[['first', 'last']] = df['fullname'].str.split(' ', expand = True)
df
```

```python
#Add data a row
df._append({'first':'Zoom'}, ignore_index=True)
```

```python
people = {
    'first': ['Tony', 'Steve'],
    'last': ['Stark', 'Rogers'],
    'email': ['IronMan@avenge.com', 'Cap@avenge.com']
}
df2 = pd.DataFrame(people)

df2
```

```python
#Append data
df = df._append(df2, ignore_index=True, sort=False)
```

```python
#Drop row specific by index
df = df.drop(index=4, axis=1)
df
```

```python
filt = df[df['last'] == 'Doe'].index
df = df.drop(index = df[filt])
df
```
